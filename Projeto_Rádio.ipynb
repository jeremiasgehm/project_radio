```
m10 = folium.Map(
    location=[-29.687162, -51.118614],
    tiles='cartodbdark_matter',
    zoom_start=14
)

mc = MarkerCluster()

for index, ouvintes in dt.iterrows():
    mc.add_child(folium.Marker([ouvintes['Lat'], ouvintes['Long']],
    popup=str(ouvintes['Quantidade']),
    tooltip=ouvintes['Bairros_NH'],
    icon=folium.Icon(icon='music'))).add_to(m10)

m10
```
